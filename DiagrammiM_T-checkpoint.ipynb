```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D 
```

# Trave su 4 appoggi: Diagrammi delle sollecitazioni

### Inviluppo dei momenti

```python
# L in m
L1 , L2, L3 = 5 , 5 , 5
q1 , q2 , q3 = 10 , 10 , 10 #CARICHI INVENTATI PER VERIFICARE CHE IL MODELLO GIRI
# Materiale acciaio
E = 200 # GPa, ossia kN/mm^2
I = 1.0e8 # mm^4
# EI = kN*m^2
EI = E * I * 1.0e-6
```

Si procede con l'algoritmo del Metodo degli Spostamenti

#### Calcolo delle sollecitazioni

```python
#Funzione che calcola le rotazioni in B e C
def calcolo_momenti_tagli(L1, L2, L3, q1, q2, q3, EI):
    #ROTAZIONI IN B e C
    A = np.array(
            [
                [(3*EI/L1 + 4*EI/L2), (2*EI/L2)],
                [(2*EI/L2), (4*EI/L2 + 3*EI/L3)],
            ]
    )

    B = np.array(
        [
            [q1*L1**2 / 8 - q2*L2**2 / 12],
            [q2*L2**2 / 12 - q3*L3**2 / 8],
        ]
    )

    x = np.linalg.solve(A, B)
    print (f"La rotazione del nodo B è pari a {x[0][0]:.5f}\nLa rotazione del nodo C è pari a {x[1][0]:.5f}")
    
    #MOMENTI SU APPOGGI
    Ma = 0
    Mba = -q1*L1**2/8 + 3*EI/L2*x[0]
    Mbc = q2*L2**2/12 + 4*EI/L2*x[0] + 2*EI/L2*x[1]
    Mcb = -q2*L2**2/12 + 2*EI/L2*x[0] + 4*EI/L2*x[1]
    Mcd = q3*L3**2/8 + 3*EI/L3*x[1]
    Md = 0
    print(f"Momento appoggio A = {np.round(Ma, 2)} kNm\nMomento appoggio B (tratto AB) = {np.round(Mba, 2)} kNm\nMomento appoggio B (tratto BC) = {np.round(Mbc, 2)} kNm\nMomento appoggio C (tratto BC) = {np.round(Mcb, 2)} kNm\nMomento appoggio C (tratto CD) = {np.round(Mcd, 2)} kNm\nMomento appoggio A = {np.round(Md, 2)} kNm\n")
    
    #TAGLI
    Ta = 3/8*q1*L1 + 3*EI/L1**2 * x[0]
    Tba = 5/8*q1*L1 - 3*EI/L1**2 * x[0]
    Tbc = q2*L2/2 + 6*EI/L2**2 * x[0] + 6*EI/L2**2 * x[1]
    Tcb = q2*L2/2 - 6*EI/L2**2 * x[0] - 6*EI/L2**2 * x[1] 
    Tcd = 5/8*q3*L3 + 3*EI/L3**2 * x[1]
    Td = 3/8 * q3*L3 - 3*EI/L3**2 * x[1]
    print(f"Taglio appoggio A = {np.round(Ta, 2)} kNm\nTaglio appoggio B (tratto AB) = {np.round(Tba, 2)} kNm\nTaglio appoggio B (tratto BC) = {np.round(Tbc, 2)} kNm\nTaglio appoggio C (tratto BC) = {np.round(Tcb, 2)} kNm\nTaglio appoggio C (tratto CD) = {np.round(Tcd, 2)} kNm\nTaglio appoggio A = {np.round(Td, 2)} kNm\n")
    return Ma, Mba, Mbc, Mcb, Mcd, Md, Ta, Tba, Tbc, Tcb, Tcd, Td
    
```

```python
calcolo_momenti_tagli(L1, L2, L3, q1, q2, q3, EI)
```

#### Diagramma del momento

Per il tracciamento del diagramma dei momenti, si rende necessario determinare l'equazione $M(z)$ lungo la trave

```python
def plot_momento(L1, L2, L3, q1, q2, q3, EI):
    Ma, Mba, Mbc, Mcb, Mcd, Md, Ta, Tba, Tbc, Tcb, Tcd, Td = calcolo_momenti_tagli(L1, L2, L3, q1, q2, q3, EI)
    #Coordinata spaziale plot
    z1 = np.linspace(0,L1, 51)
    z2 = np.linspace(0,L2, 51)
    z3 = np.linspace(0,L3, 51)
    
    #Diagramma momenti
    M_1z = q1*z1**2/2 - Ta*z1
    M_2z = q2*z2**2/2-Tbc*z2 + Mbc
    M_3z = q3*z3**2/2 - Tcd*z3 + Mcd
    
    #Plot Momenti
    plt.plot([0,L1+L2+L3],[0,0],color='black')
    plt.plot(z1, M_1z, label="Momento AB", color="red")
    plt.plot(L1 + z2, M_2z, label="Momento BC", color="red")
    plt.plot(L1 + L2 + z3, M_3z, label="Momento CD", color="red")
    
    #Fill delle aree
    plt.fill_between(z1, M_1z, color="red", alpha=0.3)  # Area sotto la curva AB
    plt.fill_between(L1 + z2, M_2z, color="red", alpha=0.3)  # Area sotto la curva BC
    plt.fill_between(L1 + L2 + z3, M_3z, color="red", alpha=0.3)  # Area sotto la curva CD
    
    
    plt.title("Bending moment diagram")
    plt.xlabel("Lunghezza (m)")
    plt.ylabel("Momento (kNm)")
    #plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plot_momento(L1, L2, L3, q1, q2, q3, EI)
```

#### Diagramma del taglio

```python
def plot_taglio(L1, L2, L3, q1, q2, q3, EI):
    Ma, Mba, Mbc, Mcb, Mcd, Md, Ta, Tba, Tbc, Tcb, Tcd, Td = calcolo_momenti_tagli(L1, L2, L3, q1, q2, q3, EI)
    #Coordinata spaziale plot
    z1 = np.linspace(0,L1, 51)
    z2 = np.linspace(0,L2, 51)
    z3 = np.linspace(0,L3, 51)
    
    #Diagramma Tagli
    T_1z = Ta - q1*z1
    T_2z = Tbc - q2*z2
    T_3z = Tcd - q3*z3
    
    #Plot Tagli
    plt.plot([0,L1+L2+L3],[0,0],color='black')
    plt.plot(z1, T_1z, label="Taglio AB", color="red")
    plt.plot(L1 + z2, T_2z, label="Taglio BC", color="red")
    plt.plot(L1 + L2 + z3, T_3z, label="Taglio CD", color="red")
    
    #Fill delle aree
    plt.fill_between(z1, T_1z, color="red", alpha=0.3)  # Area sotto la curva AB
    plt.fill_between(L1 + z2, T_2z, color="red", alpha=0.3)  # Area sotto la curva BC
    plt.fill_between(L1 + L2 + z3, T_3z, color="red", alpha=0.3)  # Area sotto la curva CD
    
    plt.title("Diagrammi dei Momenti Flettenti")
    plt.xlabel("Lunghezza (m)")
    plt.ylabel("Momento (kNm)")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plot_taglio(L1, L2, L3, q1, q2, q3, EI)
```
